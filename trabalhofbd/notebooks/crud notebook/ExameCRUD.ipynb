Import das bibliotecas necessárias


```python
import psycopg2 as pg
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
```

Configuração da conexão com o banco de dados PostgreSQL

```python
try:
  conn = pg.connect(database="projeto-fbd",host="localhost",port=5432,user='postgres',password='198019')
  print('conexão concluida')
except(Exception) as error:
  #print(error)
  print('conexão não concluida')
```

Criação de widgets


```python
#criando os widgets
try:
    tipo = widgets.Text(disbled=False, placeholder='Digite o tipo de exame', description = 'Tipo de exame', value = '')
    dia = widgets.Text(disbled=False, placeholder='Digite o dia', description = 'Dia', value = '')
    mes = widgets.Text(disbled=False, placeholder='Digite o mês', description = 'Mês', value = '')
    ano = widgets.Text(disbled=False, placeholder='Digite o ano', description = 'Ano', value = '')
    nome = widgets.Text(disbled=False, placeholder='Digite o nome', description = 'Nome', value = '')
    exame_id = widgets.Text(disbled=False, placeholder='Digite o id do exame', description = 'Id do exame', value = '')
    finalidade = widgets.Text(disbled=False, placeholder='Digite a finalidade', description = 'Finalidade', value = '')
    cpf_usuario = widgets.Text(disbled=False, placeholder='Digite o cpf do usuário', description = 'Cpf do usuário', value = '')
except(Exception) as error:
    print('erro ao criar os widgets',error)
```

Criação de botões e criação das suas funções

```python
createButton = widgets.Button(description='Criar')
readButton = widgets.Button(description='Consultar')
updateButton = widgets.Button(description='Atualizar')
deleteButton = widgets.Button(description='Deletar')
readPerIdButton = widgets.Button(description='Consultar por Id')
output = widgets.Output()


#criando as funções

def onCreateButton(b):
    try:
        tipo_value = str(tipo.value)
        dia_value = int(dia.value)
        mes_value = int(mes.value)
        ano_value = int(ano.value)
        nome_value = str(nome.value)
        exame_id_value = int(exame_id.value)
        finalidade_value = str(finalidade.value)
        cpf_usuario_value = str(cpf_usuario.value)

        cursor = conn.cursor()
        cursor.execute('insert into Exame(Tipo,Dia,Mes,Ano,Nome,ExameId,Finalidade,CpfUsuario) values(%s,%s,%s,%s,%s,%s,%s,%s)',
                       (tipo_value,dia_value,mes_value,ano_value,nome_value,exame_id_value,finalidade_value,cpf_usuario_value))
        df = pd.read_sql_query('SELECT * from Exame', conn)
        try:
            display(df)
        except:
            display('erro ao mostrar o dataframe')
    except (Exception, pg.Error) as error:
        conn.rollback()
        print('erro ao inserir os dados',error)
    finally:
        conn.commit()
        cursor.close()

def onReadButton(b):
    try:
        cursor = conn.cursor()
        cursor.execute('select * from Exame')
        df = pd.read_sql_query('SELECT * from Exame', conn)
        try:
            display(df)
        except:
            display('erro ao mostrar o dataframe')
    except (Exception, pg.Error) as error:
        conn.rollback()
        print('erro ao ler os dados',error)
    finally:
        conn.commit()
        cursor.close()
    

def onUpdateButton(b):
    try:
        tipo_value = str(tipo.value)
        dia_value = int(dia.value)
        mes_value = int(mes.value)
        ano_value = int(ano.value)
        nome_value = str(nome.value)
        exame_id_value = int(exame_id.value)
        finalidade_value = str(finalidade.value)
        cpf_usuario_value = str(cpf_usuario.value)

        cursor = conn.cursor()
        cursor.execute('update Exame set Tipo = %s, Dia = %s, Mes = %s, Ano = %s, Nome = %s, ExameId = %s, Finalidade = %s, CpfUsuario = %s where ExameId = %s',
                       (tipo_value,dia_value,mes_value,ano_value,nome_value,exame_id_value,finalidade_value,cpf_usuario_value,exame_id_value))

        df = pd.read_sql_query('SELECT * from Exame', conn)
        try:
            display(df)
        except:
            display('erro ao mostrar o dataframe')
    except (Exception, pg.Error) as error:
        conn.rollback()
        print('erro ao atualizar os dados',error)
    finally:
        conn.commit()
        cursor.close()

def onDeleteButton(b):
    try:
        cpf_usuario_value = str(cpf_usuario.value)

        cursor = conn.cursor()
        cursor.execute('delete from Exame where CpfUsuario = %s',(cpf_usuario_value,))

        df = pd.read_sql_query('SELECT * from Exame', conn)
        try:
            display(df)
        except:
            display('erro ao mostrar o dataframe')
    except (Exception, pg.Error) as error:
        conn.rollback()
        print('erro ao deletar os dados',error)
    finally:
        conn.commit()
        cursor.close()

def onReadPerIdButton(b):
    try:
        exame_id_value = int(exame_id.value)
        cursor = conn.cursor()
        cursor.execute('select * from Exame where ExameId = %s',(exame_id_value,))
        query = 'select * from Exame where ExameId = %s' % (exame_id_value) 
        df = pd.read_sql_query(query, conn)
        try:
            display(df)
        except:
            display('erro ao mostrar o dataframe')
    except (Exception, pg.Error) as error:
        conn.rollback()
        print('erro ao ler os dados',error)
    finally:
        conn.commit()
        cursor.close()
    

createButton.on_click(onCreateButton)
readButton.on_click(onReadButton)
updateButton.on_click(onUpdateButton)
deleteButton.on_click(onDeleteButton)
readPerIdButton.on_click(onReadPerIdButton)
```

Exibir widgets
